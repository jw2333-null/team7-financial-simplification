```python
!pip install sentence-transformers

from sentence_transformers import SentenceTransformer, util
import pandas as pd, numpy as np

df = pd.read_csv('dataset_model_final.csv')
model = SentenceTransformer('all-MiniLM-L6-v2')

output_cols = ['output_generic', 'output_financial', 'output_financial_t5xl', 'output_llm']
originals = df['text'].fillna('').tolist()
orig_emb = model.encode(originals)

for col in output_cols:
    outputs = df[col].fillna('').tolist()
    out_emb = model.encode(outputs)
    sims = [util.cos_sim(orig_emb[i], out_emb[i]).item() for i in range(len(originals))]
    print(f"{col}: mean={np.mean(sims):.4f} sd={np.std(sims):.4f}")
```
